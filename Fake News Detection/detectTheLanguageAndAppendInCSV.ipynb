```python
import pandas as pd
import numpy as np

# importing csv file
df = pd.read_csv('new_csv.csv')
df.head(10)
df = df.drop('Unnamed: 0',axis=1)
df.head(10)
```

```python
import nltk
import re
nltk.download(['punkt','stopwords','wordnet','crubadan']) # including nltk libraries to perfrom text language identification
from nltk.classify.textcat import TextCat
```

```python
# implementing logging mechanism to print timestamps
import logging
logging.getLogger().setLevel(logging.DEBUG)
format = "%(asctime)s: %(message)s"
logging.basicConfig(format=format, level=logging.INFO,
                        datefmt="%H:%M:%S.%F")
```

```python
def UnderstandLanguage(name, dataFrame, from_frame, to_frame):
    for i in range(from_frame, to_frame):
        print("id : ",i)
        local_copy = dataFrame['total'][i]
        logging.info("How?")
        tc = TextCat()
        language = tc.guess_language(local_copy)
        logging.info("%s", language)
        # replacing the column Language witht he detected language
        dataFrame['Language'][i] = language
        print(dataFrame)
        if i == 8000:
            dataFrame.to_csv('from_7000_8000.csv')
        elif i == 9000:
            dataFrame.to_csv('from_8000_9000.csv')
        elif i == 9999:
            dataFrame.to_csv('from_9000_10000.csv')
        
    return dataFrame
```

```python
%%time
temp_data = df.copy()
UnderstandLanguage(1,temp_data,7000,10000)
```
